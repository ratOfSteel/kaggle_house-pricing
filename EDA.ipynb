Разведочный анализ данных

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = None
```

```python
train = pd.read_csv('aux/train.csv')
```

```python
train.head(3)
```

```python
train.shape
```

1. распределение цен на дома

```python
sns.set_style("whitegrid")
sns.displot(train['SalePrice'], kind='hist');
plt.xticks(rotation=15);
plt.title('Распределение цен на дома (таргет)')
plt.xlabel('Цена')
plt.ylabel('Количество предложений')
plt.show();
```

2.  распределение числа домов по разным признакам: год постройки, ремонта и пр.

```python
df_train_num = train.select_dtypes(exclude=["object"])
```

```python
var_ind = list(range(1, 36))
plt.figure(figsize = (25, 30))
plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
for i in range(1, 36):
    plt.subplot(6, 6, i)
    plt.grid(True)
    plt.hist(x = df_train_num.iloc[: , var_ind[i - 1]], alpha = .3, color = 'green',edgecolor = "black")
    plt.title(df_train_num.columns[var_ind[i - 1]])
```
